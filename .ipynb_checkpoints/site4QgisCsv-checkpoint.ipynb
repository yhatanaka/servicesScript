```python
import sys
sys.path.append('/Users/hatanaka/venv/lib/python3.14/site-packages')
import pandas as pd
df = pd.read_csv('/Users/hatanaka/Dropbox/ジオパーク/2024_サイト再設定/鳥海山・飛島ジオパーク　サイト一覧　※2024年10月現在（最新）/Table 1-表1.csv')
df.head(16)
```

```python
# pd.set_option('display.max_rows', 100)
df_loc = df.dropna(subset=['緯度','経度'])[['2024年度No.','サイト又はその他エリア・地点','エリア','公開','緯度','経度','地形地質','自然','文化']]
df_loc.head(20)
```

```python
df_loc.dtypes
```

```python
def decideSiteType(row):
    if type(row['地形地質']) is str and type(row['自然']) is float and type(row['文化']) is float:
        return '地形地質'
    elif type(row['地形地質']) is float and type(row['自然']) is str and type(row['文化']) is float:
        return '自然'
    elif type(row['地形地質']) is float and type(row['自然']) is float and type(row['文化']) is str:
        return '文化'
    else:
        return ''
# df_geo = df_loc.apply(decideSiteType, axis=1)
df_loc['type'] = df_loc.apply(decideSiteType, axis=1)
df_export = df_loc.drop(columns=['地形地質','自然','文化'])
df_export.to_csv('/Users/hatanaka/Dropbox/ジオパーク/2024_サイト再設定/site_2024-10/2026-01-11_export.csv', index=False)
df_export.head(20)
```
